```python
#to plot bar chart of mostplayer from any country.
# From this visualization we say that aus had more players in ICC ODI ALL TIME BATSMAN'S RANKING
MostplayerCountry=df.groupby("Country").size().sort_values(ascending=False)
MostplayerCountry.head(10).plot(kind='bar')
plt.show()
```

```python
s=df.loc[0:5,['Playername','Century']]
x='Playername'
y='Century'
f=plt.figure()
f.set_figwidth(12)
f.set_figheight(10)
plt.xlabel("Playername")
plt.ylabel("Centuries")
plt.title("MAX. Centuries scored by Players in ODI")
plt.plot(x,y,data=s)
plt.show()
```

```python
manya=df.loc[(df['Century']>25)]
x='Playername'
y='Century'
f=plt.figure()
f.set_figwidth(12)
f.set_figheight(10)
plt.xlabel('Playername')
plt.ylabel('Centuries')
plt.title('Comparison Between Centuries scored by players')
plt.plot(x,y,data=manya,marker='*')
```

```python
manya=df.loc[(df['Runs']>12000)]
x='Playername'
y='Runs'
f=plt.figure()
f.set_figwidth(12)
f.set_figheight(10)
plt.xlabel('Playername')
plt.ylabel('Runs_Scored')
plt.scatter(x,y,data=manya)
plt.show()
```

```python
many=df.loc[(df['Century']>20) & (df['Country']=='IND')]
x='Playername'
y='Runs'
plt.pie(y,labels=x,data=many,autopct='% 1.1f%%')
plt.show()
```

```python
#  Here i compare The players with centuries gerater than 25 
# It show that GOD of cricket has highest number of centuries than any other in ODI Format
manya=df.loc[(df['Century']>25)]
x='Playername'
y='Century'
f=plt.figure()
f.set_figwidth(12)
f.set_figheight(10)
plt.bar(x,y,data=manya)
plt.show()
```

```python
#here we compare centuries scored by indiansin ODI as compare to rest of the world cricketers 
#and from this pie chart we know that indian cricketers contributed 21.3% of centries in the world Cricket
ab=df.loc[(df['Country']=='IND')]
centuries_by_IND =ab['Century'].sum()

cd=df.loc[(df['Country']!='IND')]
centuries_by_restofthe_worldplayers=cd['Century'].sum()
C=[centuries_by_IND,centuries_by_restofthe_worldplayers]
countries=['By_INDIANS','BY_RestOF_world']
plt.pie(C,labels=countries,autopct='% 1.1f%%')
plt.show()

```
